### 전처리를 위해 필요한 판다스와 리젝스 라이브러리 임포트

```python
import pandas as pd
import re
```

#### 전처리 할 파일 불러오기

```python
df = pd.read_csv('프로젝트 파일/강남구_단속.csv',engine='python')
```

#### 삼성로 부근, 일원로 부근등 좌표로 변환안되는 값들 추출

```python
b = re.compile(r'\w+로\s부근')
```

```python
a = df['단속장소'].str.match(b) == True
```

#### 추후 좌표지정을 위해 분류

```python
dp = df[a]
```

```python
dp
```

```python
dp.to_csv('강남_변환안됨.csv')
```

#### 좌표 반한 안되는 값들을 제외한 데이터 프레임 만들기

```python
ds = df[df['단속장소'].str.match(b) != True]
```

#### 좌표변환을 용이하게 하기 위해 단속지역과 단속장소를 하나의 컬럼으로 합침

```python
ds['단속주소'] = ds['단속지역'].str.cat(' ' + ds['단속장소'])
```

#### 합친 데이터프레임 csv파일로 export

```python
ds.to_csv('강남_단속주소.csv')
```
